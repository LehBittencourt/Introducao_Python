```python
# Exercício 1 - Crie uma função que imprima a sequência de números pares entre 1 e 20 e depois faça uma chamada à função para listar os números
lista_num = [x for x in range(0,21,2)]
print(lista_num)
```

```python
# Exercício 2 - Crie uam função que receba uma string como argumento e retorne a mesma string em letras maiúsculas. Faça uma chamada à função
def lista_string(texto):
    return print(texto.upper())
lista_string('Olá, Mundo')
```

```python
# Exercício 3 - Crie uma função que receba como parâmetro uma lista de 4 elementos, adicione 2 elementos a lista e imprima a lista"
def novaLista(lista):
    lista.append(5)
    lista.append(6)

lista = [1,2,3,4]
novaLista(lista)
print(lista)
```

```python
# Exercício 4 - Crie uma função que receba um argumento formal e uma possível lista de elementos. Faça duas chamadas à função, com apenas 1 elemento e na segunda chamada com 4 elementos"
def printVarInfo(arg1, *vartuple):
    print( 'O parâmetro passado foi:', arg1) # Imprimindo valor do primero argumento
    for item in vartuple:
        print('O parâmetro passado foi:', item) # Imprimindo valor dos outros argumentos
    return;

# Chamada da função com 1 argumento
printVarInfo(10)

# Chamada da função com 4 argumento
printVarInfo('Chocolate', 'Brigadeiro', 'Morango', 'Banana')
```

```python
# Exercício 5 - Crie uma função anônima e atribua seu retorno a uma variável chamada soma. A expressão vai receber 2 números como parâmetro e retornar a soma deles
soma = lambda n1, n2: n1+n2
print(soma(2,4))
```

```python
# Exercício 6 - Execute o código abaixo e certifique-se que compreende a diferença entre variável global e local
total = 0

def soma( arg1, arg2 ):
    total = arg1 + arg2;
    print ("Dentro da função o total é: ", total)
    return total;

soma( 10, 20 );
print ("Fora da função o total é: ", total)
```

```python
# Exercício 7 - Abaixo você encontra uma lista com temperaturas em graus Celsius, crie uma função anônima que converta cada temperatura para Fahrenheit.
# Dica: para conseguir realizar este exercício, você deve criar sua função lambda, dentro de uma função
Celsius = [39.2, 36.5, 37.3]

Fahrenheit = map(lambda x: (float(9)/5) * x + 32, Celsius)
print (list(Fahrenheit))
```

```python
# Exercício 8 - Crie uma list comprehension que imprima o quadrado dos números de 1 a 10
lista = [x**2 for x in range(1,11)]
print(lista)
```

```python
# Exercício 9 - Crie uma list comprehension que imprima as palavras com a letra 'a' no nome
palavras = ["maça", "coiote", "banana", "terreno", "Python"]

lista_palavras = [x for x in palavras if 'a' in x]
print(lista_palavras)
```

```python
# Exercício 10 - Crie uma list comprehension que imprima os números menores que 5 em um intervalo de 1 a 10
lista_numeros = [x for x in range(1,10) if x < 5]
print(lista_numeros)
```
