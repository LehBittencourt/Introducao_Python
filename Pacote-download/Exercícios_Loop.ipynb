```python
# Exercício 1 - Crie uma estrutura que pergunte qual o dia da semana. Se for domingo ou sábado, imprima na tela "Hoje é dia de descanso", caso contrário "Você precisa trabalhar!"
dia = input('Digite o dia da semana: ')
if dia == 'Domingo' or dia == 'Sábado':
    print("Hoje é dia de descanso!")
else:
    print("Você precisa trabalhar!")
```

```python
# Exercício 2 - Crie uma lista de 5 frutas e verifique se a fruta 'Morango' faz parte da lista
lista_frutas = ['Banana', 'Maçã', 'Uva', 'Morango', 'Goiaba']
for fruta in lista_frutas:
    if fruta == 'Morango':
        print('Morango está na lista de frutas!')
```

```python
# Exercício 3 - Crie uma tupla de 4 elementos, multiplique cada elemento da tupla por 2 e guarde os resultados em uma lista

# Primeira forma:
tupla_num = (2, 4, 6, 8)
lista_vazia = []

for i in tupla_num:
    x = i*2
    lista_vazia.append(x)

# Segunda forma:
tupla_num = (2, 4, 6, 8)
lista = [x*2 for x in tupla_num]

# Imprimindo as duas formas
print(lista_vazia)
print(lista)
```

```python
# Exercício 4 - Crie uma sequência de números pares entre 0 e 5 e imprima na tela
for i in range(0,5,2):
    print(i)
```

```python
# Exercício 5 - Crie uma variável chamada temperatura e atribua o valor 40. Enquanto temperatura for maior que 35, imprima as temperaturas na tela
temperatura = 40
while temperatura > 35:
    print(temperatura)
    temperatura -= 1
```

```python
# Exercício 6 - Crie uma variável contador = 0. Enquanto ela for menor que 5, imprima na tela, mas quando for encontrado o valor 3, interrompa a execução do programa
contador = 0
while contador < 5:
    if contador == 3:
        break
    print(contador)
    contador += 1
```

```python
# Exercício 7 - Crie uma lista vazia e uma variável com valor 4. Enquanto o valor da variável for menor ou igual a 20, adicione à lista, apenas os valores pares e imprima a lista
listaVazia= []
num = 4
while num <= 20:
    listaVazia.append(num)
    num += 2
print(listaVazia)
```

```python
# Exercício 10 - Faça um programa que conte quantas vezes a letra "r" aparece na frase abaixo. Use um placeholder na sua instrução de impressão
frase =  'A gratidão é a virtude das almas nobres. O quão feliz é uma pessoa depende da profundidade de sua gratidão.'
cont = 0

for caracter in frase:
    if caracter == 'r':
        cont += 1
print("O caracter r aparece %s vezes na frase." %(cont))
```
